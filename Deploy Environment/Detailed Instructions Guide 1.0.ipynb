## 2.0 Setting up your SSH key

Once you are connected to your Azure Portal ([www.portal.azure.com](http://www.portal.azure.com/)), follow the steps below to set up your SSH key.

_Note:_ _Given the audience &quot;Windows - End -User&quot; might not be familiar with Azure CLI (Command Line interface) with AZ commands and Linux/ bash I I&#39;ll showing it in Azure Portal given it&#39;s more friendly interface for whom is not a developer. In my tests I&#39;ve used Azure CLI as well, however if I add both it will be too long document, so I&#39;ll opt to focus in a simpler in a friendly interface such as Azure Portal._

You can create the SSH key following the instructions here using different ways:[_https://docs.microsoft.com/en-us/azure/virtual-machines/ssh-keys-portal_](https://docs.microsoft.com/en-us/azure/virtual-machines/ssh-keys-portal) _. The example below shows one of many ways to create it (using Azure Portal):_

### 2.2 Run ARM template (Magic link) in Azure Cloud

Now you should be prepared to run your **ARM template (Magic link)** in Azure Cloud. Follow the next steps to deploy it.

Go to the same Git repository (readme file) and click in &quot; **Deploy to Azure**&quot;, as below. That&#39;s the ARM Template link (magic link) which contain the infrastructure that will be deployed for Cube-in-a-box. The link below contains the Source-Code repository.

[https://github.com/codeindulgence/cube-in-a-box-dea.git](https://github.com/codeindulgence/cube-in-a-box-dea.git)


<img src="images\14.png" width=880 height=880 />

Click in &quot;Deploy to Azure&quot;. This will launch the deployment. However, you will need to pass the parameters needed as reference to feed the deployment automated (I mark those in &quot;yellow&quot; and described in detail below:

<img src="images\15.png" width=480 height=480 />

1. **Resource Group:** Choose the resource group that you created in the step 2.4 (&#39;cubeinbox&#39;)
2. **Region:** It&#39;s going to be the region you choose in the step 2.4 on this manual when you created the Resource Group. (&#39;_Australia East&#39;_)
3. **SSH Key:** Is the key created in step 1. Paste the whole SSH key in this field that was generate in the step 2.8 of this manual.

E.g SSH Key dummy: &#39;ssh-rsa vbk….n…..Ygm/uUcpjarM0= generated-by-azure&#39;;


1. **Instance Type:** You can choose 16Gb (Small VM) or 56Gb (large VM). _Note:_ _The larger value is going to spend more credits._
2. **SecretPassword:** Type: \&lt;choose your own secure password\&gt; (that&#39;s going to be your Jupyter password that you are going to use in the latest step to be able to launch Jupyter.
3. **My IP:** It must be your laptop&#39;s IP. Paste here your local IP address.

_Note:_ _If you don&#39;t know your IP (don&#39;t worry), just type in Google &quot;My IP&quot;, as below, or use this website:_ [https://www.whatismyip.com/](https://www.whatismyip.com/)

## 5.0 Dynamic IP addresses

**For EY only (or other Org. That uses Dynamic IP address):** Because EY network uses dynamic address space for IP allocation (which is good practice) when you restart your EY laptop your IP address might change so your EY laptop might get another IP address. If this happen when you try to reach Jupyter notebook you might get an error and you will need to check if your current IP is the same of the IP address that you used before on in Section 2.2.

For that you will need to choose &quot; **Network**&quot; tab (as below) to change your IP:

Click in every single Network interface as I am showing below (HTTP, HTTPS, SSH and AzureNotebooks) and compare it with your current IP (google your new IP &quot;My IP&quot; as you did on the section 2.2).

If it&#39;s different it means that your laptop has a different IP, so in that case you will need to update your IP. 

<img src="images\32.png" width=840 height=840 />

Change your IP (1) and save it (2) for every single Network interface (HTTP, HTTPS, SSH and AzureNotebooks) as per the example below. After that you can access your Jupyter notebook through the VM&#39;s public IP as mentioned in Section 3.0:
